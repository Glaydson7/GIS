```python
import os
import geopandas as gpd
```

```python
print("Diretório atual:", os.getcwd())

path = os.path.join(os.getcwd(), 'municipios.shp')

if os.path.exists(path):
    print("Arquivo encontrado:", path)
else:
    print("Arquivo não encontrado:", path)
```

```python
dados = gpd.read_file(path)
```

```python
dados.head()
```

```python
dados[:5].plot()
```

```python
filtro = dados['uf'] == 'MT'
```

```python
municipios_mt = dados.loc[filtro]
```

```python
municipios_mt.to_file('municipios_mt')
```

```python
municipios_mt
```

```python
municipios_filtrados = municipios_mt
```

```python
municipios_filtrados['area'] = municipios_filtrados['geometry'].area
```

```python
municipios_filtrados
```

```python
import numpy as np

municipios_filtrados['classe'] = np.where(
    municipios_filtrados['area'] < 0.1, 1, 0)

municipios_filtrados
```

```python
municipios_filtrados.to_file(os.getcwd() + '/municipios_mt.shp')
```
